## Funciones

```python
def cuadrado(n):
    return n**2

print(cuadrado(10))
```

```python
def potencia( base, exponente=2 ):
    return base**exponente

print(potencia(2, 4))
```

```python
print(potencia(2))
```

```python
def sumar(*args):
    print(args)
    total = 0
    for i in args:
        total += i
    return total


print(sumar(1,2,3,4,5,6,7,8,10))
    
```

```python
def imprimir(**kwargs):
    print(kwargs)

imprimir(a=1, b=2, c=3)
```

## Funciones de orden superior

```python
def cua(x):
    return x**2

def cub(x):
    return x**3

def aplicar(funcion, lista):
    resultado = []
    for i in lista:
        resultado.append(funcion(i))
    return resultado
    

numeros = [3,5,8]

print(aplicar(cub, numeros)) # aplicar() es una funcion de orden superior porque recibe una funcion como parametro
```

## Funciones de Orden Superior que vienen con Python 

- map
- filter
- sorted

```python
def cua(x):
    return x**2

def cub(x):
    return x**3
    
numeros = [11,5,2]
print( tuple(map(cub, numeros)) )
```

```python
anios = [1800,1840,1871,1999,2000,2020,2021]

def bisiesto(anio):
    if  anio%400==0:
        return True
    else:
        if  anio%4==0 and anio%100!=0:
            return True
    return False

print( list(filter(bisiesto, anios)) )
```

```python
def suma_acotados (*args,cota=None):
    if cota is None:
        return sum(args)
    return sum([n for n in args if -cota <= n <= cota])

print(suma_acotados(1,5,7,8,-4,cota=5))
```
