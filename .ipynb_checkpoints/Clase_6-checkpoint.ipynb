# Clase 6 - Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## üß© Consignas

1. Mostrar las **primeras 5 filas** del dataset.  
2. Calcular una nueva columna llamada `Total` = `PrecioUnitario * Cantidad`.  
3. Calcular las **ventas totales por cliente**.  
4. Determinar **qu√© producto gener√≥ mayor ingreso total**.  
5. Calcular la **cantidad total de productos vendidos por categor√≠a**.  
6. Obtener el **promedio de gasto por venta**.  
7. Filtrar las ventas de **febrero** y calcular su ingreso total.  
8. Mostrar un resumen de **estad√≠sticas descriptivas** de `PrecioUnitario`, `Cantidad` y `Total`.  
9. (Extra) Graficar:
   - Top 10 clientes por ingresos
   - Ingresos por producto (Top 10)
   - Cantidad vendida por categor√≠a
   - Distribuci√≥n del ticket (`Total`)
   - Ingresos por mes (serie temporal)

```python
df = pd.read_csv("ventas_tienda.csv", parse_dates=["Fecha"])
```

```python
df.info()
```

```python
# 1. 

df.head()
```

```python
# 2. 

df["Total"] = df["PrecioUnitario"] * df["Cantidad"]
```

```python
df
```

```python
df["Proyeccion"] = df["PrecioUnitario"] * 1.2
```

```python
df
```

```python
# 3.

ventas_por_cliente = df.groupby("Cliente")["Total"].sum().sort_values(ascending=False)
```

```python
ventas_por_cliente
```

```python
# 4.

ingresos_por_producto = df.groupby("Producto")["Total"].sum().sort_values(ascending=False)
```

```python
ingresos_por_producto.head(1)
```

```python
# 5.

cantidad_por_categoria = df.groupby("Categor√≠a")["Cantidad"].sum().sort_values(ascending=False)
```

```python
cantidad_por_categoria
```

```python
# 6.

promedio_gasto = df["Total"].mean()
```

```python
promedio_gasto
```

```python
# 7.

ventas_febrero = df[df["Fecha"].dt.month == 2]
```

```python
ventas_febrero["Total"].sum()
```

```python
df[["PrecioUnitario", "Cantidad", "Total"]].describe()
```

```python
# Top 10 clientes por ingreso 

top_clientes = ventas_por_cliente.head(10)
plt.figure()
top_clientes.plot(kind="bar")
plt.title("Top 10 clientes por ingresos")
plt.xlabel("Cliente")
plt.ylabel("Ingresos")
plt.tight_layout()
plt.show()
```

```python
# Ingresos por producto Top 10 

top_productos = ingresos_por_producto.head(10)
plt.figure()
top_productos.plot(kind="bar")
plt.title("Ingresos por producto (Top 10)")
plt.xlabel("Producto")
plt.ylabel("Ingresos")
plt.tight_layout()
plt.show()
```

```python
# Cantidad vendida por categoria

plt.figure()
cantidad_por_categoria.plot(kind="bar")
plt.title("Cantidad vendida por categor√≠a")
plt.xlabel("Categor√≠a")
plt.ylabel("Cantidad total")
plt.tight_layout()
plt.show()
```

```python
# Distribuci√≥n del ticket (Total)

plt.figure()
df["Total"].hist(bins=30)
plt.title("Distribuci√≥n del ticket (Total)")
plt.xlabel("Total")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```

```python
# Ingresos por mes

ingresos_por_mes = df.set_index("Fecha").resample("MS")["Total"].sum()
plt.figure()
ingresos_por_mes.plot(kind="line", marker="o")
plt.title("Ingresos por mes")
plt.xlabel("Mes")
plt.ylabel("Ingresos")
plt.tight_layout()
plt.show()
ingresos_por_mes
```

```python
x = np.array([0,1,2,3,4,5,6,7])
plt.plot(x, "o-r")
plt.plot(x**2, "s:g")
plt.plot(x*3, "D:b")
```

```python
x = np.array([0,1,2,3,4,5,6,7])
plt.plot(x, "o-r", label="lineal")
plt.plot(x**2, "s:g", label="cuadratica")
plt.plot(x*3, "D:b", label="triple")

plt.title("3 Funciones")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(color='c', linestyle='-', linewidth=1)
plt.minorticks_on()
plt.legend(loc="upper left")
```

## Tipos de Graficos

| Tipo de Gr√°fico                          | Caracter√≠sticas                                                                                     | Uso principal                                                   | Ejemplo pr√°ctico                                                                          |
| ---------------------------------------- | --------------------------------------------------------------------------------------------------- | --------------------------------------------------------------- | ----------------------------------------------------------------------------------------- |
| **Gr√°fico de barras**                    | Barras rectangulares (verticales u horizontales). Permite comparar valores de distintas categor√≠as. | Comparar cantidades entre categor√≠as discretas.                 | Ventas de productos A, B y C en un trimestre.                                             |
| **Gr√°fico de l√≠neas**                    | L√≠nea continua que une puntos de datos en el tiempo.                                                | Analizar tendencias o evoluci√≥n temporal.                       | Evoluci√≥n del precio del d√≥lar en un a√±o.                                                 |
| **Gr√°fico circular (torta/pie chart)**   | C√≠rculo dividido en porciones proporcionales.                                                       | Mostrar distribuci√≥n porcentual de un todo.                     | Distribuci√≥n del gasto mensual: 40% vivienda, 30% comida, 20% transporte, 10% ocio.       |
| **Gr√°fico de dispersi√≥n (scatter plot)** | Puntos en un plano cartesiano que representan pares de variables.                                   | Identificar correlaciones o relaciones entre dos variables.     | Relaci√≥n entre horas de estudio y nota de examen.                                         |
| **Histograma**                           | Barras agrupadas en intervalos que muestran frecuencia.                                             | Analizar la distribuci√≥n de datos continuos.                    | Distribuci√≥n de edades de clientes en un supermercado.                                    |
| **Boxplot (diagrama de caja y bigotes)** | Caja con mediana, cuartiles y valores at√≠picos.                                                     | Identificar la dispersi√≥n y outliers en datos num√©ricos.        | Distribuci√≥n de salarios en una empresa.                                                  |
| **√Årea (Area chart)**                    | Similar al gr√°fico de l√≠neas, pero con el √°rea bajo la curva coloreada.                             | Mostrar la evoluci√≥n de una magnitud acumulada.                 | Consumo energ√©tico mensual acumulado de un hogar.                                         |
| **Gr√°fico de radar (ara√±a)**             | Varias dimensiones representadas en ejes radiales.                                                  | Comparar m√∫ltiples variables de un objeto.                      | Comparaci√≥n de habilidades de candidatos: comunicaci√≥n, liderazgo, an√°lisis, creatividad. |
| **Mapa de calor (heatmap)**              | Matriz de valores codificada en colores.                                                            | Visualizar patrones e intensidad en grandes vol√∫menes de datos. | Temperaturas promedio por hora y d√≠a de la semana en una ciudad.                          |
| **Diagrama de dispersi√≥n con burbujas**  | Similar al scatter plot, pero con tama√±o de burbuja representando una tercera variable.             | Analizar tres variables simult√°neamente.                        | Relaci√≥n entre ventas (eje X), beneficio (eje Y) y tama√±o de mercado (tama√±o de burbuja). |


```python
# Scatter - Grafico de puntos

x,y = np.random.random((2,30))
c = np.random.randint(0,9,30)
s = np.random.randint(20,250,30)

plt.scatter(x,y,c=c,s=s) # s= size, c = color
```

```python
# Histograma - Distribuci√≥n de numeros, se conforma de barras cuyo alto es la
#              cantidad de datos que estan en el segmento comprendido por la barra

x1 = np.random.normal(4,3,1000)
x2 = np.random.normal(1,1,1000)

plt.hist((x1,x2), bins=20, label=["Grupo 1", "Grupo 2"])
plt.title("Histograma")
plt.legend()
```

```python
x1 = np.random.normal(4,3,1000)
x2 = np.random.normal(1,1,1000)
plt.hist((x1,x2), bins=10, label=["Grupo 1", "Grupo 2"], histtype="stepfilled", alpha=0.5)
plt.title("Histograma")
plt.legend()
```

```python
# Barras

grupos = ['Ni√±os', 'Jovenes', 'Adultos', 'Ancianos']
hombres = [28,45,34,88]
mujeres = [10,72,37,65]

indice = np.arange(len(grupos))

## Se crean las primeras barras
plt.bar(indice, hombres, label='Hombres')

## Se crean las segundas barras y se apilan sobre las primeras
plt.bar(indice, mujeres, label='Mujeres',  bottom=hombres)

plt.xticks(indice, grupos)
plt.ylabel("Participanes")
plt.xlabel("Grupos")
plt.title('Participantes por g√©nero')
plt.legend()

plt.show()
```

```python
price = [39.5,39.9,45.4,38.9,33.34]

# Definicion de look & feel
plt.rcParams['font.sans-serif'] =[u'Arial']
plt.rcParams['axes.unicode_minus'] = False

# Dibujo
plt.barh(range(5), price, align = 'center',color='steelblue', alpha = 0.8)
# Agregar etiqueta de eje
plt.xlabel('precio')
# A√±adir t√≠tulo
plt.title('Comparaci√≥n de precios de un libro en diferentes plataformas')
# Agregar etiquetas de marca
plt.yticks(range(5),['Amazonas','Ateneo','Distal','Cuspide','Libros'])
# Establecer el rango de escala del eje X
plt.xlim([32,47])

# Agregue etiquetas num√©ricas a cada gr√°fico de barras
for x,y in enumerate(price):
    plt.text(y,x,'%s' %y,va='center')
    
# Mostrar gr√°ficos
plt.show()
```

```python
# Creaci√≥n de un array de valores distribuidos uniformemente
uniform = np.arange(-100, 100)

# Creaci√≥n de un array de distribuci√≥n de valores normales
normal = np.random.normal(size = 100)*30

fig, (ax1, ax2) = plt.subplots(nrows = 1,
                               ncols = 2,
                               figsize =(9, 4),
                               sharey = True)

ax1.set_title('Distribuci√≥n Uniforme')
ax1.set_ylabel('Valores Observados')
ax1.violinplot(uniform)

ax2.set_title('Distribuci√≥n Normal')
ax2.violinplot(normal)

plt.show()
```
