## Aggregate

```python
data.groupby("Sex").Age.agg(["max", "min", "mean", "median", "count"])
```

```python
data.groupby("Sex").Age.agg(
    sex_max=("max"),
    sex_min=("min")
)
```

```python
def categorizar(x):
    m = x.mean()
    return True if m > 29 else False

data.groupby("Sex").Age.agg(["max", "min", "mean", categorizar])
```

```python
data.describe(percentiles=[.1,.25,.5,.9])
```

```python
data.groupby("Sex").Age.agg(["max", "min", "mean", lambda x: True if x.mean() > 29 else False ])
```

Importar el archivo GoldPrice.csv (el mismo de la clase, del ltimo apunte Pandas) como dataframe y realizar los siguientes clculos:
1. Crear una nueva columna que solo contenga el ao y llamarla Year.
2. Determinar el promedio de precios del oro por ao.
   
1. Graficar el promedio del valor del oro por ao con un grfico de barras.
2. Calcular el ao con mayor promedio y el ao con menor promedio.

```python
gold.index.year
```

```python
gold["Year"] = gold.index.year
```

```python
gold
```

```python
gold[["Year", "Price"]].groupby("Year").mean()
```

```python
gold[["Year", "Price"]].groupby("Year").mean().plot.bar()
```

```python
max_promedio = float(gold[["Year", "Price"]].groupby("Year").mean().max())
min_promedio = float(gold[["Year", "Price"]].groupby("Year").mean().min())


idx_max_promedio = int(gold[["Year", "Price"]].groupby("Year").mean().idxmax())
idx_min_promedio = int(gold[["Year", "Price"]].groupby("Year").mean().idxmin())

```

```python
print(f"El ao de mayor promedio fue {idx_max_promedio} con un promedio de {round(max_promedio,2)}, El ao de menor promedio fue {idx_min_promedio}, con un promedio de {round(min_promedio,2)}")
```
