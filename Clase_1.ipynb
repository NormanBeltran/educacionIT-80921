```python
# crear una lista de 30 nombres aleatorios utilizando la libreria random y el metodo choice
import random
nombres = ["juan", "ana", "pedro", "maria", "lucia", "carlos", "sofia", "diego", "laura", "jose"]
nombres_aleatorios = []
for i in range(30):
    nombres_aleatorios.append(random.choice(nombres))   
```
